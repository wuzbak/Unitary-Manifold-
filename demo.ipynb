## 1 · CMB Observables — Prediction vs Observation

A single integer pair **(n₁, n₂) = (5, 7)** chosen by 5D Kaluza-Klein topology predicts three independent CMB observables.  
All three currently pass observational constraints.

| Observable | Framework prediction | Measurement | Status |
|---|---|---|---|
| Spectral index *nₛ* | 0.9635 | 0.9649 ± 0.0042 (Planck 2018) | ✅ 0.33σ |
| Tensor-to-scalar *r* | 0.0315 | < 0.036 (BICEP/Keck 2021) | ✅ satisfied |
| Birefringence *β* | 0.331° | [0.22°, 0.38°] admissible window | ✅ in window |

````python
from src.core.braided_winding import braided_ns_r
from src.core.inflation import PLANCK_NS_CENTRAL, PLANCK_NS_SIGMA
from src.core.falsification_check import (
    BETA_PRIMARY, BETA_SHADOW, BETA_MIN, BETA_MAX,
    BETA_GAP_LO, BETA_GAP_HI, BETA_THEORY_UNCERTAINTY
)

# ── Framework predictions ────────────────────────────────────────────────────
pred = braided_ns_r(5, 7)          # (n1=5, n2=7) braided KK winding modes

NS_PRED   = pred.ns                # predicted spectral index
R_PRED    = pred.r_eff             # predicted tensor-to-scalar ratio
BETA_PRED = BETA_PRIMARY           # predicted birefringence (primary sector)
CS_SPEED  = pred.c_s               # braided sound speed
K_CS      = pred.k_cs             # Chern-Simons level

# ── Observational benchmarks ─────────────────────────────────────────────────
NS_OBS    = PLANCK_NS_CENTRAL      # 0.9649
NS_SIGMA  = PLANCK_NS_SIGMA        # 0.0042
R_LIMIT   = 0.036                  # BICEP/Keck 95% CL

ns_tension = abs(NS_PRED - NS_OBS) / NS_SIGMA

print("══════════════════════════════════════════════════════")
print(" Unitary Manifold — CMB Predictions from (n₁,n₂)=(5,7)")
print("══════════════════════════════════════════════════════")
print(f"  Chern-Simons level k_CS  = {K_CS}  (= 5² + 7²  resonance identity)")
print(f"  Braided sound speed c_s  = {CS_SPEED:.6f}  (= 12/37)")
print()
print(f"  nₛ predicted : {NS_PRED:.4f}    observed: {NS_OBS:.4f} ± {NS_SIGMA:.4f}  ({ns_tension:.2f}σ)")
print(f"  r  predicted : {R_PRED:.4f}    limit   : < {R_LIMIT}  {'✅ passes' if R_PRED < R_LIMIT else '❌ fails'}")
print(f"  β  predicted : {BETA_PRED:.3f}°    window  : [{BETA_MIN}°, {BETA_MAX}°]  ✅ in window")
print("══════════════════════════════════════════════════════")
````

````python
# ── Visual: prediction vs observation comparison chart ───────────────────────
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
fig.suptitle(
    'Unitary Manifold CMB predictions from a single integer pair (n₁, n₂) = (5, 7)',
    fontsize=12, fontweight='bold'
)

GREEN  = '#2ca02c'
BLUE   = '#1f77b4'
ORANGE = '#ff7f0e'
GREY   = '#d3d3d3'

# ── Panel A: spectral index nₛ ───────────────────────────────────────────────
ax = axes[0]
ax.bar(['Planck 2018\nobservation', 'UM\nprediction'],
       [NS_OBS, NS_PRED],
       color=[GREY, GREEN], edgecolor='black', width=0.5)
ax.errorbar([0], [NS_OBS], yerr=[NS_SIGMA], fmt='none', color='black', capsize=6, lw=2)
ax.set_ylim(0.940, 0.985)
ax.set_ylabel('nₛ')
ax.set_title(f'Spectral index nₛ\n({ns_tension:.2f}σ tension)')
ax.axhline(NS_OBS, color='black', lw=0.8, ls='--', alpha=0.5)
for i, v in enumerate([NS_OBS, NS_PRED]):
    ax.text(i, v + 0.0003, f'{v:.4f}', ha='center', va='bottom', fontsize=9)

# ── Panel B: tensor-to-scalar ratio r ───────────────────────────────────────
ax = axes[1]
ax.bar(['BICEP/Keck\n95% limit', 'UM\nprediction'],
       [R_LIMIT, R_PRED],
       color=[ORANGE, GREEN], edgecolor='black', width=0.5, alpha=0.85)
ax.set_ylim(0, 0.055)
ax.set_ylabel('r')
ax.set_title(f'Tensor-to-scalar ratio r\n(prediction {R_PRED:.4f} < limit {R_LIMIT})')
ax.annotate('limit', xy=(0, R_LIMIT), xytext=(0, R_LIMIT + 0.003),
            ha='center', fontsize=9, color=ORANGE)
for i, v in enumerate([R_LIMIT, R_PRED]):
    ax.text(i, v + 0.001, f'{v:.4f}', ha='center', va='bottom', fontsize=9)

# ── Panel C: birefringence β window ─────────────────────────────────────────
ax = axes[2]
window_span = BETA_MAX - BETA_MIN
ax.barh(['Admissible\nwindow'], [window_span], left=[BETA_MIN],
        color=GREY, edgecolor='black', height=0.4, alpha=0.6, label='window')
ax.barh(['Forbidden\ngap'], [BETA_GAP_HI - BETA_GAP_LO], left=[BETA_GAP_LO],
        color='#d62728', edgecolor='black', height=0.4, alpha=0.6, label='forbidden gap')
ax.axvline(BETA_PRED, color=GREEN, lw=2.5, label=f'UM β = {BETA_PRED:.3f}°')
ax.axvline(BETA_SHADOW, color=BLUE, lw=1.5, ls='--', label=f'shadow β = {BETA_SHADOW:.3f}°')
ax.set_xlabel('β (degrees)')
ax.set_title('Birefringence β\n(LiteBIRD test ~2032)')
ax.legend(fontsize=8, loc='lower right')

plt.tight_layout()
plt.show()
print(f'\nAll three observables: nₛ ({ns_tension:.2f}σ) | r (passes) | β (in window)')
````

## 2 · 5D → 4D Projection: KK Radion Field Evolution

The KK radion φ is the scalar field encoding the size of the compact fifth dimension at each point.  
The Walker–Pearson integrator evolves (g_μν, B_μ, φ) on a 1D spatial grid using RK4.  
Watching φ(x) evolve shows the extra dimension stabilising to its background value φ₀ = 1.

*This is what Pillar 2 (field evolution) looks like when you run it interactively.*

````python
from src.core.evolution import FieldState, run_evolution, information_current
from src.core.metric import compute_curvature

# ── Initialise flat 5D background ───────────────────────────────────────────
N, dx = 64, 0.1
state = FieldState.flat(N=N, dx=dx, rng=np.random.default_rng(42))

print("5D field initialisation")
print(f"  Grid points   N = {N},  spacing dx = {dx}")
print(f"  Initial φ     mean = {state.phi.mean():.6f}  std = {state.phi.std():.2e}")
print(f"  Initial |B|   mean = {np.linalg.norm(state.B, axis=1).mean():.2e}")
print()

# ── Evolve 50 RK4 steps ─────────────────────────────────────────────────────
dt, steps = 1e-3, 50
history = run_evolution(state, dt=dt, steps=steps)

print(f"Evolved {steps} RK4 steps  (dt = {dt},  final t = {history[-1].t:.4f})")
````

````python
x = np.arange(N) * dx

# ── Collect φ mean and Ricci scalar at each step ─────────────────────────────
times   = [s.t for s in history]
phi_mean = [s.phi.mean() for s in history]
phi_std  = [s.phi.std()  for s in history]

R_means = []
for s in history:
    _, _, _, R = compute_curvature(s.g, s.B, s.phi, s.dx)
    R_means.append(float(R.mean()))

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle(
    '5D → 4D Projection: KK radion φ evolution (Walker–Pearson RK4 integrator)',
    fontsize=12, fontweight='bold'
)

# ── Left: φ(x) snapshots ────────────────────────────────────────────────────
ax = axes[0]
snap_indices = [0, steps // 3, 2 * steps // 3, steps]
cmap = plt.cm.viridis
colors = [cmap(i / (len(snap_indices) - 1)) for i in range(len(snap_indices))]
for idx, col in zip(snap_indices, colors):
    t_label = f't = {history[idx].t:.3f}'
    ax.plot(x, history[idx].phi, color=col, label=t_label, lw=1.5)
ax.axhline(1.0, color='black', lw=0.8, ls='--', alpha=0.5, label='φ₀ = 1 (background)')
ax.set_xlabel('x (spatial grid)')
ax.set_ylabel('φ (KK radion)')
ax.set_title('φ(x) at four time slices')
ax.legend(fontsize=8)

# ── Centre: φ mean ± std over time ──────────────────────────────────────────
ax = axes[1]
ax.plot(times, phi_mean, color=GREEN, lw=2, label='⟨φ⟩')
ax.fill_between(times,
                np.array(phi_mean) - np.array(phi_std),
                np.array(phi_mean) + np.array(phi_std),
                alpha=0.2, color=GREEN, label='±σ')
ax.axhline(1.0, color='black', lw=0.8, ls='--', alpha=0.5, label='φ₀ = 1')
ax.set_xlabel('time')
ax.set_ylabel('⟨φ⟩')
ax.set_title('Radion mean ± std over time\n(extra dimension stabilising)')
ax.legend(fontsize=9)

# ── Right: Ricci scalar R over time ─────────────────────────────────────────
ax = axes[2]
ax.plot(times, R_means, color=BLUE, lw=2)
ax.set_xlabel('time')
ax.set_ylabel('⟨R⟩ (Ricci scalar)')
ax.set_title('4D Ricci scalar from 5D projection\n(geometric curvature sourced by φ)')
ax.axhline(0, color='black', lw=0.8, ls='--', alpha=0.4)

plt.tight_layout()
plt.show()
````
